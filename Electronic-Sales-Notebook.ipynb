```python
for col in dataframe.columns:
    pct_missing = dataframe[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
# Dropping null values
dataframe = dataframe.dropna()
```

```python
# Checking for duplicated values
dataframe.duplicated().sum()
```

```python
# Removal of duplicated values
dataframe = dataframe.drop_duplicates()
```

```python
# Checking for duplicated column names
id, date, address = 0,0,0
for i in range(0, len(dataframe)):
    check_id = dataframe['Order ID'].iloc[i]
    check_date = dataframe['Order Date'].iloc[i]
    check_add = dataframe['Purchase Address'].iloc[i]
    if check_id == 'Order ID':
        id += 1
    if check_date == 'Order Date':
        date += 1
    if check_add == 'Purchase Address':
        address += 1   
print("No of columns where column name is copied in the Order ID column:", id)
print("No of columns where column name is copied in the Order Date column:", date)
print("No of columns where column name is copied in the Purchase Address column:", address)
```

```python
# Removal of column names and its duplicates
dataframe = dataframe[dataframe['Order Date'] != 'Order Date']
dataframe
```

```python
# Recheck of column names as an entry in the dataset
id, date, address = 0,0,0
for i in range(0, len(dataframe)):
    check_id = dataframe['Order ID'].iloc[i]
    check_date = dataframe['Order Date'].iloc[i]
    check_add = dataframe['Purchase Address'].iloc[i]
    if check_id == 'Order ID':
        id += 1
    if check_date == 'Order Date':
        date += 1
    if check_add == 'Purchase Address':
        address += 1   
print("No of columns where column name is copied in the Order ID column:", id)
print("No of columns where column name is copied in the Order Date column:", date)
print("No of columns where column name is copied in the Purchase Address column:", address)
```

```python
# Changing data types to accurate types
dataframe['Order ID'] = pd.to_numeric(dataframe['Order ID'])
dataframe['Quantity Ordered'] = pd.to_numeric(dataframe['Quantity Ordered'])
dataframe['Price Each'] = pd.to_numeric(dataframe['Price Each'])
dataframe['Order Date'] = pd.to_datetime(dataframe['Order Date'], format='mixed')

dataframe
```

```python
# Creating a column for sales of products
dataframe['Sales'] = (dataframe['Price Each'] * dataframe['Quantity Ordered']).round(2)
```

```python
# Creating columns related to date-time
dataframe['Day'] = dataframe['Order Date'].dt.day
dataframe['Month'] = dataframe['Order Date'].dt.month
dataframe['Week Day'] = dataframe['Order Date'].dt.day_name()
dataframe['Hour'] = dataframe['Order Date'].dt.hour

dataframe
```

```python
# Creating a column for time of the day
times_of_day = []
for i in dataframe['Hour']:
    if i>=5 and i <= 12 :
        times_of_day.append('Morning')
    elif i> 12 and i <= 18:
        times_of_day.append('Afternoon') 
    elif i> 18 and i <= 22:
        times_of_day.append('Evening')
    else:
        times_of_day.append('Night')
dataframe['Times of Day'] = times_of_day

dataframe
```

```python
# Functions created to split address and get the city and state part of it
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
```

```python
# Creating columns for city and state
dataframe['City'] = dataframe['Purchase Address'].apply(lambda x: get_city(x))
dataframe['State'] = dataframe['Purchase Address'].apply(lambda x: get_state(x))

dataframe
```

```python
# Dropping Purchase Address and Order Date columns since they are processed
dataframe.drop(['Order Date', 'Purchase Address'], axis = 1, inplace= True)
```

```python
#Check the Dimensions
shape=dataframe.shape
print('Number of Rows : ',shape[0])
print('Number of Columns : ',shape[1])
```

```python
# Saving the processed dataset
dataframe.to_csv("Processed_sales_data.csv", index = False)
```
